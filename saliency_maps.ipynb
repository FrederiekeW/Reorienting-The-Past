## Saliency Maps

**Overview over generating example saliency maps and saliency map overlays for figure 1**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread(r"C:\...\Analysis\Objects\All\27_t_0.png") #change image per saliency map
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert image so it can be processed
saliency = cv2.saliency.StaticSaliencyFineGrained_create()

(success, saliency_map) = saliency.computeSaliency(image)

# Convert saliency to 0–255, because you have to convert image type aka most png expect 8-bit images 
#basically the brightness scale, so 0 = black and 255 = white, so we turn the normal 0-1 to a normal image brightness value
# Why 255?  2^8^= 256 , so 255 values of brightnesss
saliency_map = (saliency_map * 255).astype("uint8")

# Save saliency map alone
cv2.imwrite(r"C:\...\Analysis\Objects\All\saliencymap_27.png", saliency_map) #remember renaming file
```

```python
# Convert saliency to heatmap colors
heatmap = cv2.applyColorMap(saliency_map, cv2.COLORMAP_JET)

# Overlay with original image
overlay = cv2.addWeighted(image, 0.6, heatmap, 0.4, 0)

cv2.imwrite(r"C:\...\Analysis\Objects\All\s_overlay_27.png", overlay) #remember renaming file
```
